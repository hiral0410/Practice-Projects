### Outliers Checking:

```python
collist=df.columns.values
ncol=30
nrows=14
plt.figure(figsize=(ncol,3*ncol))
for i in range (0,len(collist)):
    plt.subplot(nrows,ncol,i+1)
    sns.boxplot(y=df[collist[i]],color='green',orient='v')
    plt.tight_layout()
```

From the above outcome we can see that columns['Citric Acid','Alcohol','Quality'] has less outliers and columns['Fixed Acidity','Volatile Acidity','Residual Sugar','Chlorides','free sulfur dioxiode','total sulfur dioxide','Density','pH','Sulphates'] has more outliers.
